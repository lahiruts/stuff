```python
#select sensitive hidden units

model='exp_pdnn/dnn6_fbank_new/'
num_speakers=40
num_phones=2008
max_spk_entropy= num_speakers * -(1.0/num_speakers)*log2(1.0/num_speakers)
max_phn_entropy= num_phones * -(1.0/num_phones)*log2(1.0/num_phones)

layer=1


phn_sens_name = model+'pdfSensitivitylayer'+str(layer)
spk_sens_name = model+'spkSensitivitylayer'+str(layer)

#print phn_sens_name
#print spk_sens_name

#process pdf sensitivity
phn_sens_file = open(phn_sens_name,'r')
phn_vals={}
phn_index=0
for line in phn_sens_file:
    phn_vals[max_phn_entropy - float(line.rstrip())]=phn_index
    phn_index += 1
    #if index > 1023:
    #    break


phn_keys = sorted(phn_vals.keys(),reverse=True)
print len(phn_keys)

phn_sens_units=[]
for i in range(100):
    phn_sens_units.append(phn_vals.get(phn_keys[i]))
    
phn_np_array = np.asarray(sorted(phn_sens_units,reverse=True))
phn_sens_file.close()
#process speaker senstivity

spk_sens_file = open(spk_sens_name,'r')
spk_vals={}
spk_index=0
for line in spk_sens_file:
    spk_vals[max_spk_entropy - float(line.rstrip())]=spk_index
    spk_index += 1
    #if index > 1023:
    #    break


spk_keys = sorted(spk_vals.keys(),reverse=True)
print len(spk_keys)

spk_sens_units=[]
for i in range(100):
    spk_sens_units.append(spk_vals.get(spk_keys[i]))
    
spk_np_array=np.asarray(sorted(spk_sens_units,reverse=True))
spk_sens_file.close()

np.intersect1d(phn_np_array, spk_np_array)
```
