```python
import numpy as np
import math

def calc_entropy(activationVector):
    entropy=0.0
    summation = np.sum(activationVector)
    activationVector = activationVector / summation
    for val in activationVector:
        entropy = entropy - val * math.log(val,2)
    return entropy

base_dir='sensitivity/activations/dev_0330/'
speakers = '050 050 051 052 053 22g 22h 420 421 422 423'
channels = 'wv1'
noises = 'clean car babble restaurant street airport train'

layer_size=2048

speakerList=speakers.split()
channelList = channels.split()
noisesList = noises.split()

for channel in channelList:
    for noise in noisesList:
        for layer in range(1,8):
            print '-------------'+channel+'--------- ' + noise + ' ------------------' + layer + '--------'
            spk_counts={}
            spk_vectors={}

            for speaker in speakerList:
                
                frame_count=0
                activation_file = open(base_dir + speaker + '/'+channel+'/'+noise +'/'+str(layer)'_activations.ark','r')
                for line in activation_file:
                    act_line = (line.rstrip()).split()
                    if len(act_line) == 2:
                                utt1= act_line[0]
                                #print ' started reading hidden activations for ' + act_line[0]
                    elif len(act_line) == layer_size + 1:
                        frame_count =frame_count + 1
                        if speaker in spk_vectors.keys():
                            spk_vectors[speaker] = np.add(np.array(act_line[:layer_size],dtype=np.float),spk_vectors.get(speaker))
                        else:
                            spk_vectors[speaker] = np.array(act_line[:layer_size],dtype=np.float)
                    else:
                        frame_count =frame_count + 1
                        if speaker in spk_vectors.keys():
                            spk_vectors[speaker] = np.add(np.array(act_line,dtype=np.float),spk_vectors.get(speaker))
                        else:
                            spk_vectors[speaker] = np.array(act_line,dtype=np.float)
                spk_counts[speaker]=frame_count


            #average phoneme vectors
            for spk in spk_counts.keys() :
                spk_vectors[spk] = spk_vectors.get(spk)/spk_counts.get(spk)

            weight_array=np.zeros(layer_size)
            for spk in spk_vectors.keys():
                weight_array=np.vstack((weight_array,spk_vectors.get(spk)))

            weight_array = np.delete(weight_array,0,0)

            num_speakers = len(weight_array)
            max_entropy= num_speakers * -(1.0/num_speakers)*math.log(1.0/num_speakers, 2)
            print max_entropy
            #print np.sum(weight_array[1:,0])
            ent_val=[]
            for i in range(layer_size):
                ent_val.append(calc_entropy(weight_array[:,i]))
                
            total_ent=0.0
            sens_val={}
            node_idx=0
            for ent in ent_val :
                sens = (max_entropy - ent)/max_entropy
                total_ent = total_ent + sens
                sens_val[node_idx]=sens
                node_idx = node_idx + 1

            print total_ent/layer_size
            
            sens_units=[]
            for key in sorted(sens_val,key=sens_val.get,reverse=True):
                sens_units.append(key)
                
            np.savetxt(base_dir+'speakerSensitivity_%s_%s_%d'%(channel,noise,layer),sens_val.values())
            np.savetxt(base_dir+'spkUnits_%s_%s_%d'%(channel,noise,layer),sens_units,fmt='%4i')
```
